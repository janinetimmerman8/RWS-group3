```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

dataframes = [results_1, results_2, results_3, results_4]

subplot_titles = ['K-means-distance', 'K-means-travel time', 'Simulated annealing', 'Frequency-based']

# Make the subplots
fig, axes = plt.subplots(2, 2, figsize=(20, 16), subplot_kw={'title': ''})
fig.suptitle('4 Heatmap Plots of Probability of <18 mins')

# Set the day order
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Set the time order
time_range_order = ['0-5', '6-9', '10-14', '15-18', '19-23']

for i, df in enumerate(dataframes):
    ax = axes[i // 2, i % 2]
    ax.set_title(subplot_titles[i])  

    df['Time ranges'] = pd.Categorical(df['Time ranges'], categories=time_range_order, ordered=True)

    full_data = pd.DataFrame(index=time_range_order, columns=day_order, dtype=float)
    full_data = full_data.fillna(0)  
    
    for index, row in df.iterrows():
        full_data.at[row['Time ranges'], row['Day']] = float(row['Probability of <18 mins'])
    
    sns.heatmap(full_data, annot=True, fmt=".2%", cmap='viridis', cbar=True, ax=ax)
    ax.set_xlabel('Day')
    ax.set_ylabel('Time ranges')
plt.savefig('heatmap_data.png', dpi=300)
plt.show()
```
